**2.1. Exploring your data** 

Think about your pandas functions for checking out a dataframe. 

```python
# copy the original dataframe 
df = PROPERTIES_ORIG.copy()
# set columns to be easier to type
df.columns = [x.lower().replace(' ', '_').replace('&', '').replace('__', '_') 
              for x in df.columns]
# retain apparent place name codes in a dict
CODES_DICT = df.iloc[0].to_dict()
df = (df.drop(index=[0]).rename(columns={'unnamed:_0':'date'}))
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(by=['date']).reset_index(drop=True)
df['date'] = df['date'].dt.strftime('%Y-%m-%d')

# for 'unnamed' columns, drop them if entirely null, otherwise raise error
cols = [x for x in df.columns if 'unnamed' in x]
for col in cols:
    m = df[col].notnull()
    assert not df[m].shape[0], col
df = df.drop(columns=cols)

# set float types for house prices
for col in [x for x in df.columns if x != 'date']:
    df[col] = df[col].astype(float)

# reshape and add back in the city codes
df = df.set_index('date').T.reset_index().rename(columns={'index':'city'})
df['city_code'] = df['city'].map(CODES_DICT)
df = df.set_index(['city', 'city_code']).reset_index()

print(df.shape)
df.head(5)
```

>There are only 32 London boroughs, but there are 45 rows in this dataframe. Some of these rows apparently need to be consolidated. I got a list of the 32 names from Wikipedia:

```python
# borough names from Wikipedia
boroughs = ['City of London', 'City of Westminster', 'Kensington and Chelsea', 
            'Hammersmith and Fulham', 'Wandsworth', 'Lambeth', 'Southwark', 
            'Tower Hamlets', 'Hackney', 'Islington', 'Camden', 'Brent', 'Ealing', 
            'Hounslow', 'Richmond upon Thames', 'Kingston upon Thames', 'Merton', 
            'Sutton', 'Croydon', 'Bromley', 'Lewisham', 'Greenwich', 'Bexley', 
            'Havering', 'Barking and Dagenham', 'Redbridge', 'Newham', 
            'Waltham Forest', 'Haringey', 'Enfield', 'Barnet', 'Harrow', 'Hillingdon']
boroughs = [x.lower().replace(' ', '_') for x in boroughs]
print(boroughs)
```

>...and below I'm going to use fuzzy string matching to determine which of these should be consolidated:

```python
# use fuzzywuzzy to compare string similarity. 
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

new_names = dict()
# for every name in the dataframe 
for br in df['city'].tolist():
    # find the closest "proper" name i.e. find closest name from the Wikipedia list 
    best_match = process.extractOne(br, boroughs)
    # if it is an exact match, add to the new_name mapping 
    if best_match[1] == 100:
        new_names[br] = best_match[0]
    else:
        # otherwise, will have to handle manually
        print(br, best_match)
```

```python
# manually update new_names dict where it is obvious
new_names['barking_dagenham'] = 'barking_and_dagenham'
new_names['hammersmith_fulham'] = 'hammersmith_and_fulham'
new_names['kensington_chelsea'] = 'kensington_and_chelsea'
new_names['westminster'] = 'city_of_westminster'
new_names['inner_london'] = 'city_of_london'
new_names['outer_london'] = 'city_of_london'
new_names['london'] = 'city_of_london'
```

```python
# display closest match for remaining values of city from dataframe
for br in [x for x in df['city'] if x not in new_names.keys()]:
    best_match = process.extractOne(br, boroughs)
    print(br, best_match)
```

```python
df['city_new_name'] = np.nan
df['city_new_name'] = df['city'].map(new_names)
```

> Some of the borough names are duplicated, we'll group by the updated borough name and return the mean value for each, in order to consolidate, and we'll reshape this into a tidy dataset:

```python
cols = [x for x in df.columns if x.startswith('19') or x.startswith('20')]
gr = (df.dropna(subset=['city_new_name']).groupby('city_new_name')[cols].mean()
         .sort_values(by=cols[-1]).T.reset_index(drop=False))
gr.head()
```

> Let's use `groupby` again to translate these into observations by year:

```python
gr['date'] = pd.to_datetime(gr['date'])
gr['year'] = gr['date'].dt.year
gr = gr.drop(columns=['date']).groupby('year').mean()
gr
```

```python
msg = 'Should have all years in the index'
assert gr.index.tolist() == list(range(1995, 2025)), msg
```

> Let's plot all of the values by year:

```python
w = 17
fig, ax = plt.subplots(figsize=(w, w/1.618))

for c in gr.columns:
    ax.plot(gr.index, gr[c], label=c)
plt.grid()
plt.legend()
```

> We'll plot the 2004-2024 increase in a horizontal bar plot, which can answer the primary question of this task:

```python
w = 10
fs = (w, w/1.618)
percent_increase = (((gr.loc[2024] - gr.loc[2004]) / gr.loc[2004])
                        .sort_values(ascending=True))

fig, ax = plt.subplots(figsize=fs)
ax.barh(percent_increase.index, percent_increase.values)

for i,v in enumerate(percent_increase.values):
    ax.text(v, i, f"{100*v:.1f}%", va='center', ha='left')
    
ax.set_xlim(left=0, right=2)
ticks = np.linspace(0, 2, num=10)
xlabs = [f"{100*x:.1f}%" for x in ticks]
plt.xticks(rotation=45, ticks=ticks, labels=xlabs)
plt.xlabel('Percent Increase, 2004-2024')
plt.show()
```

```python
# display a formatted summary table of the data
summary = gr.loc[[2004, 2024], :].T
summary = summary.map(lambda x: f"{x/1000:.3f}K")
summary['percent_increase'] = percent_increase
summary['percent_increase'] = summary['percent_increase'].map(lambda x: f"{100*x:.1f}%")
summary = summary.reindex(percent_increase.index)
summary.iloc[::-1, :].head(10)
```
